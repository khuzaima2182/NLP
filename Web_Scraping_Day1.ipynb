**Importing the Libraries**

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

**Putting in the URL**

```python
URL = "https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue"
page = requests.get(URL)
soup = BeautifulSoup(page.text, 'html')
print(soup)
```

**Finding out the Table**

```python
soup.find_all('table')[0]
```

```python
soup.find('table', class_ = 'wikitable sortable')
```

```python
table = soup.find_all('table')
```

```python
titles = soup.find_all('th')
```

```python
print(titles)
```

```python
data = [title.get_text().strip() for title in titles]
desired_data = data[1:9]
print(desired_data)
```

**Converting it into the DataFrame**

```python
df = pd.DataFrame(columns = desired_data)
df
```

**Now Scrapping the data into the rows and column**

```python
column_data = soup.find_all('tr')
print(column_data)
```

```python
for row in column_data[2:52]:
  row_data = row.find_all('td')
  row_table_data = [dataa.text.strip() for dataa in row_data]

  length = len(df)
  df.loc[length] = row_table_data
df
```

```python
df.drop('State-owned', axis=1, inplace=True)
```

```python
df.drop('Ref.', axis=1, inplace=True)
```
